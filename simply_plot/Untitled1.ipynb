# 1. Average speed per country

```python
import simply_plot
import pandas

data = pandas.read_csv(
    '/Users/marcusflodihn/Desktop/Data Warehousing/assignments/2/SpeedPerCountry.csv',
    sep=';'
)

converted_data = list()

for value in data['average speed']:
    converted_data.append(value.replace(',', '.'))

options = {
    'title': 'Average speed (km/h) per country ',
    'x_axis_name': 'Country',
    'y_axis_name': 'Average speed(km/h)',
    'data': converted_data,
    'labels': data['country'],
    'bar_width': 5,
    'bar_spacing': 0.5,
    'size': (19,6)
}

plt.ylim(ymin=60, ymax=100)

chart = simply_plot.BarChart(**options)
```

# Further examining Nuth by year
